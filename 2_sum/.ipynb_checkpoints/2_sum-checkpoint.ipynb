```python
import pandas as pd

data = pd.read_csv('data.txt', header=None)[0].tolist()
stream = set(data)
```

```python
targets = list(range(-10000,10001))
target_found = set()
for counter, number in enumerate(stream):
    for  target in targets:
        if (target-number in stream) and (number!=target-number):
            target_found.add(target)
            targets.remove(target)
            print ('progress is at: '+str(round(counter*100/len(stream),4))+'%')
print('Total targets found: ', len(target_found))


```
