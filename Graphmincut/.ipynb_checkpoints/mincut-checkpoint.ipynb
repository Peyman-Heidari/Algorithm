```python
import random, math
import pandas as pd
import numpy as np
from itertools import compress, chain
from operator import itemgetter
```

```python
data = pd.read_csv('data.csv', header=None)
```

```python
edges = []
for node in range(0,data.shape[0]):
    for j in range(1,data.shape[1]):
        if not np.isnan(data.loc[node,j]):
            pair = (node+1,data.loc[node,j])
            edges.append(pair)
```

```python
def primer(x):
    duplicate = [1]*len(x)
    for first in range(len(x)):
        for second in range(first, len(x)):
            if (x[first][0]==x[second][1]) & (x[first][1]==x[second][0]):
                duplicate[second] = 0
    unique_x = list(compress(x, duplicate))
    return unique_x
```

```python
def remove_self_edge(x):
    self = [1]*len(x)
    for edge in range(len(x)):
        if x[edge][0]==x[edge][1]:
            self[edge] = 0
    clean_x = list(compress(x, self))
    return clean_x
```

```python
def node_unifier(x):
    m= len(x)
    edge_number = max(list(chain(*x)))+1
    unified_edge = random.randint(0,m-1)
    A = x[unified_edge][0]
    B = x[unified_edge][1]
    x.remove(x[unified_edge]) 
    for edge in range(len(x)):
        if (x[edge][0]==A) or (x[edge][0]==B):
            x.insert(edge, (edge_number,x[edge][1]))
            x.remove(x[edge+1])
        
        if (x[edge][1]==A) or (x[edge][1]==B):
            x.insert(edge, (x[edge][0],edge_number))
            x.remove(x[edge+1])   
    unified_x = remove_self_edge(x)
    return unified_x
            
```

```python
def cutter_finder(x):
    if len(set(list(chain(*x))))==2:
        return len(x)
    else:
        while len(set(list(chain(*x))))!=2:
            x = node_unifier(x)
        print "min crossing in this iteration is: ", len(x)
        return len(x)
            
```

```python
def min_cross_cutter(x):
    n = len(x)
    no_iterations = int(math.ceil(math.log(n)*n))
    min_crossing = n**2
    for iteration in range(1, no_iterations):
        random.seed(iteration*100)
        crossings = cutter_finder(x)
        print "iteration", float(iteration)/float(no_iterations)
        if crossings < min_crossing: 
            min_crossing = crossings
    print "************************************"
    return "Final min crossing is", min_crossing
        
```

```python
unique_edges = primer(edges)
```

```python
min_cross_cutter(unique_edges)
```
