```python
def inversion(x):
    inv = 0
    merged = []
    n = len(x)
    if n<2:
        return x, 0
    else:
        left = x[:int(n/2)]
        right = x[int(n/2):]
        left, inv_left = inversion(left)
        right, inv_right = inversion(right)
        i = 0
        j = 0
        n_left = len(left)
        n_right = len(right)
        while i < n_left and j <n_right : 
            if left[i] < right[j]:
                merged.append(left[i])
                i +=1
            elif left[i] > right[j]:
                merged.append(right[j])
                j +=1
                inv = inv+n_left-i
        merged += left[i:]
        merged += right[j:]
        sum_inv = inv_left+inv_right+inv
        return merged, sum_inv
```

```python
inversion([1, 20,10,12,3,0])
```

```python
import pandas as pd
data = pd.read_csv('data.csv', header=None)
data.columns = ['numbers']
list_data = data['numbers'].tolist()
list_data
```

```python
d,n= inversion(list_data)
```
